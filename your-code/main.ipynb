```python
# Defining the function based on the previous step:

def unif_rand_numb(a, b, n):
    
    return uniform.rvs(loc=a, scale=b-a, size=n)
```

```python
result1 = unif_rand_numb(10, 15, 100)
print (result1)
```

```python
result2 = unif_rand_numb(10, 60, 1000)
print (result2)
```

```python
n_bins = 10  # number of bins

plt.hist(result1, bins=n_bins, alpha=0.5, label="Dist 1")
plt.hist(result2, bins=n_bins, alpha=0.5, label="Dist 2")

plt.xlabel("Value")
plt.ylabel("Count")
plt.legend()
plt.title("Uniform Distributions")

plt.show()
```

Getting superimposed plots, which does not make things very clear.

For this reason, we will split the histogram results into two distinct ones.

```python
plt.subplot(1, 2, 1)    # 1 row (only one horizontal line); 2 columns (split into 2 plots - left and right); 1 - first position (left)
plt.hist(result1, bins=n_bins)
plt.ylim(0, 120)   # forcing same scale
plt.title("Distribution 1")

plt.subplot(1, 2, 2)   # i.e., the right plot
plt.hist(result2, bins=n_bins)
plt.ylim(0, 120)   # forcing same scale
plt.title("Distribution 2")

plt.show()
```

To control scale steps on both graphs in order to get expected result for the histograms as above:

```python
plt.subplot(1, 2, 1)
plt.hist(result1)
plt.ylim(0, 120)   # forcing same scale
plt.title("Distribution 1")
#plt.xticks(np.arange(10, 16, 1))  # step for left plot is 1 by 1

plt.xticks(np.arange(int(min(result1)), int(max(result1))+1, 1))   ## int(max(result1))+1 is to ensure the last number is included due to the nature of "arange"

plt.subplot(1, 2, 2)
plt.hist(result2)
plt.ylim(0, 120)   # forcing same scale
plt.title("Distribution 2")
#plt.xticks(np.arange(0, 51, 10))  # step for right plot is 10 by 10

plt.xticks(np.arange(int(min(result2)), int(max(result2))+1, 10))   # int(max(result2))+1 is to ensure the last number is included due to the nature of "arange"
```
