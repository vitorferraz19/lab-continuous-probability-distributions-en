How are the two distributions different?

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.

1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# Defining the function:

def normal_rand_numb(mean, std, size):
    
    return norm.rvs(loc=mean, scale=std, size=size)
```

```python
result3 = normal_rand_numb(10, 1, 1000)
print(result3)
```

```python
result4 = normal_rand_numb(10, 50, 1000)
print(result4)
```

```python
n = 30  # bins

# Create figure with shared Y-axis
fig = plt.figure()

# Plot 1 (std = 1)
plt.subplot(1, 2, 1)
plt.hist(result3, bins=n)
plt.title("Mean=10, Std=1")


# Plot 2 (std = 50)
plt.subplot(1, 2, 2)
plt.hist(result4, bins=n)
plt.title("Mean=10, Std=50")


plt.show()
```

How are the two distributions different?

```python
# Both distributions are centered at the same mean (10), but the one with a small standard deviation (1) is tightly concentrated, forming a tall, narrow peak with values close to 10, while the one with a large standard deviation (50) is widely spread, producing a flatter, broader shape with values extending far from the mean. The key insight is that the standard deviation controls how dispersed the data is—small values concentrate data near the mean, while large values spread it out significantly.
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
def exp_rand_num(mean, size):
    return np.random.exponential(scale=mean, size=size)
```

```python
result5 = exp_rand_num(10, 10)
```

```python
result6 = exp_rand_num(10, 100)
```

```python
n = 100  # bins

# Getting counts:
counts1, _ = np.histogram(result5, bins=n)
counts2, _ = np.histogram(result6, bins=n)

ymax = max(counts1.max(), counts2.max())

# Left plot
plt.subplot(1, 2, 1)
plt.hist(result5, bins=n)
plt.ylim(0, ymax)
plt.title("Size = 10")

# Right plot
plt.subplot(1, 2, 2)
plt.hist(result6, bins=n)
plt.ylim(0, ymax)
plt.title("Size = 100")

plt.show()
```

How are the two distributions different?

```python
# The two distributions differ mainly in how smooth and representative they are due to the sample size. With a small sample (size = 10), the histogram appears irregular and noisy, making it harder to clearly see the exponential shape. With a larger sample (size = 100), the distribution becomes smoother and more clearly shows the expected exponential pattern, which is right-skewed with a high frequency of small values and a long tail to the right.
```

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
from scipy.stats import expon

# This is same as saying P(x<15):

mean = 10
z = 15

prob_less15min = expon.cdf(z, scale=mean)
print(prob_less15min)
```

What is the probability that the customer will spend more than 15 minutes

```python
# Given that P(X < z) + P(X > z) = 1:

# P(X > 15) = 1 - P(X < 15). Hence:

p_over15min = 1 - prob_less15min
print(p_over15min)

# Note that, for continuous distributions: 
# P(X < z) and P(X ≤ z) are no different
```
